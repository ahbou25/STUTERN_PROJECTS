```python
# An inner merge operation between the orders table and products table
products_orders = df_order.merge(df_products, how='inner', indicator = True)
products_orders.head()

```

Observation:
he inner merge operation connected the two datasets together into one, and aligned the rows from each based on common attributes or columns. 

```python
#Performing another inner merge operation with the customers table using the resulting table
df_all = products_orders.merge(df_customers, how='inner')
df_all.head()

```

Observation: It returned only those rows that have common values

```python
# Perform an inner merge operation between customers and orders tables. 
customer_orders = pd.merge(df_customers, df_order, on = 'customer_id', how = 'inner')
customer_orders

```

Observations: The inner merge operation between the orders table and the customers table returns all the rows from orders and customer tables where the customer_id are equal.

```python
#Performing an outer merge between customers and products tables. 
customer_products = pd.merge(df_customers,df_products, how = 'outer')
customer_products
```

```python
#Perform an outer merge between the customers and products tables. using left_index,right_index
customers_products = pd.merge(df_customers, df_products, left_index=True, right_index=True, how ='outer', indicator = True)
customers_products.head()
```

Observation: The customers and products dataframes only merged when the matching indexes as 'TRUE' (left_index, right_index), the customers and products dataframes were able to merge.

```python
#Performing a left merge between orders and customers tables. 

orders_customers = pd.merge(df_order,df_customers, how='left', on='customer_id')
orders_customers
```

Observation: The left merge operation between the prducts and the customers table returns all the rows from the orders table and only the rows from the customer table where the customer_id on both tables are equal. Also has NaN representing empty spaces in the dataframe

 

```python
#Performing a right merge between product and customers tables
products_customers = pd.merge(df_products, df_customers, left_index=True, right_index=True, how = 'right')
products_customers.head()
```

```python
#Using the pandas concatenation method to join the three datasets together
df_concat = pd.concat([df_order, df_customers,df_products], axis=1)
df_concat
```
