```python
# Importing libraries
import matplotlib 
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('anscombe_quartet_data.csv')
df
```

```python
#To get the correlation coefficient of anscombe's dataset
df.corr()
```

```python
#To create subplots that show how each x column correlates with each y-axis column

#First approach
fig = plt.figure(figsize=(16, 10))
x = [0,15]
y = [0,10.7]
x1 = df['x1']
y1 = df['y1'] 
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color ="green")
plt.plot(x,y, color = 'black', linestyle = 'solid')

x = [0,15]
y = [0, 10.34]
x2 = df['x2']
y2 = df['y2'] 
plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color ="green")
plt.plot(x,y, color = 'black', linestyle = 'solid')

x = [0,15]
y = [0, 13.8]
x3 = df['x3']
y3 = df['y3'] 
plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color ="green")
plt.plot(x,y, color = 'black', linestyle = 'solid')

x = [0,18]
y = [0,13.6]
x4 = df['x4']
y4 = df['y4'] 
plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color ="green")
plt.plot(x,y, color = 'black', linestyle = 'solid')

plt.subplots_adjust(wspace=0.5, hspace =0.5, bottom =0.1, top =0.9)
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16, 10))
x = [0,12]
y = [0, 10.84]
ax1.scatter(x1, y1, color = "red")
ax1.plot(x,y, color = 'black')
ax1.set_title('(x1,y1)')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')

x = [0,12]
y = [0, 9.24]
ax2.scatter(x2, y2, color = "red")
ax2.plot(x,y, color = 'black')
ax2.set_title('(x2,y2)')
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')

x = [0,12]
y = [0, 11.9]
ax3.scatter(x3, y3, color = "red")
ax3.plot(x,y, color = 'black')
ax3.set_title('(x3,y3)')
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')

x = [0,12]
y = [0, 12.9]
ax4.scatter(x4, y4, color = "red")
ax4.plot(x,y, color = 'black')
ax4.set_title('(x4,y4)')
ax4.set_xlabel('x4')
ax4.set_ylabel('y4')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```
