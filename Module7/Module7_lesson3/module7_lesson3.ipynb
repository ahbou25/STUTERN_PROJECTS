```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
company_sales_data = pd.read_csv('company_sales_data.csv')
company_sales_data.head()
```

```python
# Comparing any two products sales data using a bar chart.
fig = plt.figure(figsize=(12, 8))

plt.bar([month-0.25 for month in company_sales_data['month_number']], company_sales_data['toothpaste'], label = 'tooth_paste', width= 0.50)
plt.bar([month+0.25 for month in company_sales_data['month_number']], company_sales_data['facewash'], label = 'Face_Wash', width= 0.50)

plt.xticks(company_sales_data['month_number'])
plt.title('toothpaste and facewash sales data')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.show()
```

```python
# Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.
fig = plt.figure(figsize=(12, 8))

average_monthly_sales = [10000 for i in range(len(company_sales_data['month_number']))]
plt.bar(company_sales_data['month_number'], company_sales_data['bathingsoap'])
plt.plot(company_sales_data['month_number'], average_monthly_sales, linewidth=4.5)

plt.xticks(company_sales_data['month_number'])
plt.title('Bathing soap sales data on average in monthly data sales')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.show()
```

```python
#The total sales data for the year of each product and represent it on a pie chart
fig = plt.figure(figsize=(12, 8))

sales_data   = [company_sales_data ['facecream'].sum(), company_sales_data ['facewash'].sum(), company_sales_data ['toothpaste'].sum(), 
                company_sales_data ['bathingsoap'].sum(), company_sales_data ['shampoo'].sum(), company_sales_data ['moisturizer'].sum()]
labels = ['Face cream', 'Fase wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(sales_data, labels=labels, autopct='%0.1f%%')
plt.title('Company Sales')
plt.legend(loc='upper left')
plt.show()
```
