```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('house_pricing_dataset.csv')
data.head()
```

```python
# House price distribution
fig = plt.figure(figsize=(12, 8))

number_bins = 500
plt.hist(data['price'], number_bins)
plt.xscale("log")
plt.title('House Price Distribution \n')
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average house prices across the top 10 cities
fig = plt.figure(figsize=(12, 8))

avg_house_price_by_city = data['price'].groupby(data['city']).mean().sort_values(ascending=False)

plt.hist(avg_house_price_by_city.head(10))
plt.title('Average House Price Across Top 10 Cities Distribution \n')
plt.xlabel('Average Sales Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Trend of house prices from the earliest date to the latest date in the dataset
fig = plt.figure(figsize=(12, 8))

plt.scatter(data['date'], data['price'])
plt.title('Trend of house prices from earliest date to the latest date \n')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(data['date'], rotation = 'vertical')

plt.show()
```

```python
# Distribution of maximum bedroom size across the top 10 cities
fig = plt.figure(figsize=(12, 8))

max_bedroom_by_city = data['bedrooms'].groupby(data['city']).max().sort_values(ascending=False)

plt.hist(max_bedroom_by_city.head(10))
plt.title('Maximum Bedroom Size Across Top 10 Cities Distribution \n')
plt.xlabel('Bedroom Size')
plt.ylabel('Frequency')

plt.show()
```

```python
# The distribution of oldest houses across top 10 cities
fig = plt.figure(figsize=(12, 8))

oldest_houses_by_city = data['yr_built'].groupby(data['city']).max().sort_values(ascending=True)

plt.hist(oldest_houses_by_city.head(10))
plt.title('Oldest Houses Across Top 10 Cities Distribution \n')
plt.xlabel('Year Build')
plt.ylabel('Frequency')

plt.show()
```

```python
# The distribution of the years the houses were last renovated
fig = plt.figure(figsize=(12, 8))

plt.hist(data[data['yr_renovated'] != 0]['yr_renovated'], bins=50)
plt.title('Years Houses was Last Renovated Distribution \n')
plt.xlabel('Year Renovated')
plt.ylabel('Frequency')

plt.show()
```

```python
# The distribution of average basement square feets across top 10 cities
fig = plt.figure(figsize=(12, 8))

avg_basement_square_feets_by_city = data['sqft_basement'].groupby(data['city']).mean().sort_values(ascending=False)
plt.hist(avg_basement_square_feets_by_city.head(10))
plt.title('Average Basement Square Feets Across Top 10 Cities Distribution \n')
plt.xlabel('Average Basement Square Feets')
plt.ylabel('Frequency')
plt.show()
```

```python
# The relationship between prices and number of bedrooms
fig = plt.figure(figsize=(12, 8))

plt.scatter(data['bedrooms'], data['price'])
plt.title('Relationship Between Prices and Number of Bedrooms \n')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')

plt.show()
```

```python
# the relationship between number of bedrooms and year the house was built
fig = plt.figure(figsize=(12, 8))

plt.scatter(data['bedrooms'], data['yr_built'])
plt.title('Relationship Between number of bedroom and year house was built \n')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Year build')

plt.show()


```
