```python
import matplotlib 

from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv("company_sales_data.csv")
print(df)
```

```python

plt.plot(df.month_number, df.facecream, color='black', linestyle='--', marker='o')
plt.plot(df.month_number, df.facewash, color='blue', linestyle=':', marker='s')
plt.plot(df.month_number, df.toothpaste, color='yellow', linestyle=':', marker='*')
plt.plot(df.month_number, df.bathingsoap, color='red', linestyle='--',marker='o')
plt.plot(df.month_number, df.shampoo, color='purple', linestyle=':', marker='s')
plt.plot(df.month_number, df.moisturizer, color='brown', linestyle=':',marker='*')

plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'], loc ="upper left")
plt.xlabel("Months")
plt.ylabel("Toiletries")
plt.title("Company Sales Data")
plt.show()
```
